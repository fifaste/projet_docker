```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = 'dm_office_sales.csv'
df = pd.read_csv(url)
df
```

```python
sns.scatterplot(data = df, x = "salary", y = "sales")
```

```python
plt.figure(figsize = (12,8))
sns.scatterplot(data = df, x = "salary", y = "sales")
```

```python
sns.scatterplot(data = df, x = "salary", y = "sales", hue='division')
```

```python
sns.scatterplot(data = df, x = "salary", y = "sales", hue='work experience')
```

```python
sns.scatterplot(data = df, x = "salary", y = "sales", hue='division', palette='viridis')
```

```python
sns.scatterplot(data = df, x = "salary", y = "sales", style='work experience')
```

```python
plt.figure(figsize = (12,8))
sns.scatterplot(data = df, x = "salary", y = "sales", style='level of education', hue='level of education', s=1000)
```
